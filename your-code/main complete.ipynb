# Challenge 2 - Advanced Functions

In this part of the lab, we would like to test complex conditions on the entire dataset at once. Let's start by finding the number of rows where the CGPA is higher than 9 and the student has conducted research.

```python
# your code here

admissions_df3 = pd.read_csv("Admission_Predict.csv")
counter = 0

def complex_con(row):
    global counter
    if (row['CGPA'] > 9) & (row['Research'] == 1):
        counter = counter + 1
    else:
        pass

admissions_df3.apply(complex_con, axis=1)
print(counter)
```

Now return all the rows where the CGPA is greater than 9 and the SOP score is less than 3.5. Find the mean chance of admit for these applicants.

```python
# your code here

def challenge2(row):
    if (row['CGPA'] > 9) & (row['SOP'] < 3.5):
        return "Yes"
    else:
        return "No"

admissions_df3["Challenge2"] = admissions_df3.apply(challenge2, axis = 1) # Creating a filter

Challenge2_df = admissions_df3[admissions_df3['Challenge2'] == 'Yes'] #filtering in new df 
Challenge2_df.columns = Challenge2_df.columns.str.strip() # Deleting index columns empty spaces, leading and trailing

Challenge2_df['Chance of Admit'].mean()
```

We would like to create a deciding factor column for each student. We standardize several columns and then pick the most important factor from a lookup table. If the standardized value is above 0.8, the student will be accepted.

We will start by creating a standardized column for `CGPA`, `GRE Score` and `LOR`. We will name these columns `CGPA_std`, `GRE_std`, and `LOR_std` respecively.

Recall that standardizing a column is done by subtracting the mean of the column from all observations in the column and then dividing each observation in the column by the column's standard deviation.

In the cell below, write a standardization function (a function that takes a column as input and returns a standardized column as output). Make sure to use the `numpy` versions of mean and standard deviation.

```python
def standardize(col):
#     print(col)
#     print(col.mean())
#     print(col.std())
    return (col-col.mean())/col.std()
```

Now create the standardized columns `CGPA_std`, `GRE_std`, and `LOR_std` and add them to the `admissions` dataframe.

```python
# your code here

admissions_df['CGPA_std'] = standardize(admissions_df['CGPA'])
admissions_df['GRE_std'] = standardize(admissions_df['GRE Score'])
admissions_df['LOR_std'] = standardize(admissions_df['LOR'])
admissions_df
```

We will generate the decision choice at random using the code below. Please run the cell.

```python
# Libraries
from random import choices
```

```python
std_columns = ['CGPA_std', 'GRE_std', 'LOR_std']

decision_choice = choices(std_columns, k=admissions_df.shape[0])
```

Now create the deciding column using the `lookup` function. The lookup column is `decision_choice` found above. Call the column resulting from the lookup function `deciding_column` and add it to the `admissions` dataframe.

```python
# your code here

admissions_df['Choice'] = decision_choice
deciding_list = []
g=-1

for i in decision_choice:
    g=g+1
    if (i == 'GRE_std'):
        deciding_list.append((admissions_df.at[g, 'GRE_std']))
    elif (i == 'CGPA_std'):
        deciding_list.append((admissions_df.at[g, 'CGPA_std']))
    else:
        deciding_list.append((admissions_df.at[g, 'LOR_std']))

admissions_df['deciding_column'] = deciding_list
admissions_df
```

Create a column called `decision` in the `admissions` dataframe. Assign 1 to this column if the value of `deciding_column` is greater than 0.8 and 0 otherwise.

```python
# your code here

decision_list = []

for i in admissions_df['deciding_column']:
    if i > 0.8:
        decision_list.append(1)
    else:
        decision_list.append(0)
        
admissions_df['decision'] = decision_list
admissions_df
```

How many applicants will be accepted to the program using the decision column? Compute the result below.

```python
# your code here

applicants_df = admissions_df[admissions_df['decision'] == 1]
applicants_df
```

To increase our coding efficiency, let's make a number of changes to our dataframe in one line of code.

In the cell below, remove all non character symbols from the column names, replace all spaces in column names to underscores, and change all upper case characters to lower case. Assign these new column values to `admissions.columns`.

```python
# your code here

admissions_df.columns = (admissions_df.columns.str.replace(' ','_').str.lower())
admissions_df.columns = admissions_df.columns.str.replace('.','')
admissions_df
```

In the cell below, give all student with a university rating of 4 or higher a 10 point boost on their GRE score and split the column into 4 bins using the `cut` function. Assign this new score to the variable `adjusted_gre`.

```python
# your code here

# creating a list with ratings equal/higher than 4 and boosting their GRE_score +10 and creating new column

gre_boost_list = []
g=-1

for i in admissions_df['university_rating']:
    g=g+1
    if i >= 4:
        gre_boost_list.append((admissions_df.at[g, 'gre_score'])+10)
    else:
        gre_boost_list.append(admissions_df.at[g, 'gre_score'])

admissions_df['adjusted_gre'] = gre_boost_list
display(admissions_df)

#Creting 4 bins
category = ["Bin Low", "Bin Medium", "Bin High", "Bin Very High"]
bins = pd.cut(admissions_df['adjusted_gre'],4, labels = category)
bins.value_counts()
```
